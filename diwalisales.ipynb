```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\88017\OneDrive\Desktop\Python\Diwali Sales Data.csv", encoding = "unicode_escape")
```

```python
# Removing unnecessary columns

df = df.drop(["Status","unnamed1"], axis = 1)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df.columns
```

```python
df.rename(columns = {'Cust_name': 'Customer_name'}, inplace=True)
```

```python
df[['Orders', 'Amount', 'User_ID']].describe()
```

```python
Orders_by_Gender = df[["Orders","Gender"]].groupby("Gender").sum("Orders")
```

```python
# Number of order placed by gender 

Orders_by_Gender.plot(kind= 'bar', title ="Number of Items Ordered", ylabel='Order Item')
```

```python
# Number of order placed by gender

ax = sns.countplot(x= 'Gender', data = df)
plt.title('Number of Orders by Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
df.columns
```

```python
#Amount summed up by gender

Total_amount = df[["Gender","Amount"]].groupby("Gender").sum("Amount")
az = sns.barplot(x="Gender",y="Amount",data= Total_amount )

for bar in az.containers:
    az.bar_label(bar)

plt.show()
```

```python
df[["Gender","Amount"]].groupby("Gender").sum("Amount")
```

```python
# Count of orders by gender & age group

sns.countplot(x="Age Group", data=df, hue="Gender")
plt.show()
```

```python
# Purchase by age group 

ag = df[["Age Group","Amount"]].groupby("Age Group").sum("Amount").sort_values(by = "Amount", ascending=False)
ap = sns.barplot(x="Age Group", y="Amount", data=ag)

plt.ylabel("Amount in million")
plt.show()
```

```python
# Top 10 state by sales

ag = df[["State","Amount"]].groupby("State").sum("Amount").sort_values(by = "Amount", ascending=False).head(10)
ap = sns.barplot(x="State", y="Amount", data=ag,hue= "State", palette='husl')

sns.set(rc = {'figure.figsize' :(15,5)})

plt.grid(axis = 'x')
plt.ylabel("Amount in million")
plt.show()
```

```python
# Top 10 state by Number of Orders

ag = df[["State","Orders"]].groupby("State").sum("Orders").sort_values(by = "Orders", ascending=False).head(10)
ap = sns.barplot(x="State", y="Orders", data=ag, hue= "State", palette='husl')

sns.set(rc = {'figure.figsize' :(15,5)})

for bars in ap.containers:
    ap.bar_label(bars)

plt.title("No Of Ordered Items")
plt.grid(which = 'minor', axis = 'x')
plt.show()
```

```python
df['Marital_Status'] = df['Marital_Status'].astype(str)
```

```python
# Order By Marital status

az = df[["Marital_Status", "Orders"]].groupby("Marital_Status").sum().sort_values(by = "Orders", ascending = False)
sns.barplot(x = 'Marital_Status', y = 'Orders', data = az)

plt.figure(figsize=(15, 5))
plt.show()
```

```python
df.columns
```

```python
df["Marital_Status"] = df["Marital_Status"].replace('0', 'Married')
```

```python
df["Marital_Status"] = df["Marital_Status"].replace('1', 'Single')
```

```python
# Shopping by occupation

ag = df[["Occupation","Orders"]].groupby("Occupation").sum("Orders").sort_values(by = "Orders", ascending=False)
ap = sns.barplot(x="Occupation", y="Orders", data=ag, palette='husl', hue = "Occupation" )

sns.set(rc = {'figure.figsize' :(15,5)})

for bars in ap.containers:
    ap.bar_label(bars)

plt.title("No Of Ordered Items")
plt.grid(which = 'minor', axis = 'x')
plt.show()
```

```python
# Sales by product category

ag = df[["Product_Category","Orders"]].groupby("Product_Category").sum("Orders").head(10)
ap = sns.barplot(x="Product_Category", y="Orders", data=ag, palette='husl', hue = "Product_Category" )

sns.set(rc = {'figure.figsize' :(15,5)})

for bars in ap.containers:
    ap.bar_label(bars)

plt.title("No Of Ordered Items")
plt.grid(which = 'minor', axis = 'x')
plt.show()
```
